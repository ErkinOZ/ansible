# Data on which the model was not trained

```python
df2.columns
```

```python
x_df2 = df2[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population',
       'Income per Room'
]]
```

```python
x_df2
```

```python
df2['Predicted Price'] = lm.predict(x_df2)
```

```python
df2
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(df2['Price'], df2['Predicted Price'])
mse = mean_squared_error(df2['Price'], df2['Predicted Price'])
rmse = mean_squared_error(df2['Price'], df2['Predicted Price'], squared=False)
r2 = r2_score(df2['Price'], df2['Predicted Price'])

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
df2['Error'] = df2['Price'] - df2['Predicted Price']
```

```python
df2
```

```python

df2['Error %'] = abs(df2['Error']) / df2['Price'] * 100

def classify_error(error_pct):
    if error_pct < 5:
        return "Accurate" 
    elif error_pct < 15:
        return "Moderate" 
    else:
        return "Poor"  

df2['Prediction Class'] = df2['Error %'].apply(classify_error)
```

```python
df2
```

```python
print(df2['Prediction Class'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(8,5))
sns.histplot(df2['Error %'], bins=30, kde=True)
plt.axvline(x=5, color='g', linestyle='--', label='5% (Accurate)')
plt.axvline(x=15, color='r', linestyle='--', label='15% (Moderate)')
plt.xlabel("Prediction Error (%)")
plt.ylabel("Count")
plt.title("Distribution of Prediction Errors")
plt.legend()
plt.show()
```
